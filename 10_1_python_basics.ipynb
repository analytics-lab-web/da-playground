```python
import pandas as pd

# 1️⃣ CSV 불러오기
df = pd.read_csv("tips.csv")

# 2️⃣ 결측치 확인
df.isnull().sum()  
# → 각 열(Column)마다 NaN의 개수를 보여줌

# 3️⃣ 결측치 제거
df_drop = df.dropna()  
# → NaN이 포함된 행(Row)을 통째로 삭제

# 4️⃣ 결측치 채우기
df_fill = df.fillna(0)  
# → NaN을 0으로 채움

# 5️⃣ 결측치 채우기 (평균으로)
df['tip'] = df['tip'].fillna(df['tip'].mean())

# 6️⃣ 중복 제거
df_unique = df.drop_duplicates()

# 7️⃣ 그룹별 평균 구하기
grouped = df.groupby('sex')['tip'].mean()
print(grouped)

# 8️⃣ 여러 기준으로 그룹화
grouped_multi = df.groupby(['sex', 'smoker'])['tip'].mean()
print(grouped_multi)

# 9️⃣ 조건 필터링 예시
high_tip = df[df['tip'] > 6]  
# → 팁이 6달러 이상인 행만 선택
```
