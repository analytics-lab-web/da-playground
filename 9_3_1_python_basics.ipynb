```python
# ë“±ê¸‰ íŒì • í•¨ìˆ˜: í‰ê·  ì ìˆ˜ë¥¼ ë°›ì•„ ë“±ê¸‰(A~F) ë¬¸ìì—´ ë°˜í™˜
def grade_from_avg(avg):                     # avg: í‰ê·  ì ìˆ˜(0~100 ê°€ì •)
    # ì¡°ê±´ë¬¸ì€ ìœ„ì—ì„œ ì•„ë˜ë¡œ í‰ê°€ë˜ë¯€ë¡œ, ë†’ì€ ì ìˆ˜ë¶€í„° ê²€ì‚¬
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    elif avg >= 60:
        return "D"
    else:
        return "F"

# ê°„ë‹¨ í…ŒìŠ¤íŠ¸
for a in [95, 84, 73, 62, 20]:
    print(a, "â†’", grade_from_avg(a))
```
